# Normal Form Matrix Game: Stag Hunt Game (Sandholm 2012, 6-7)
* In the Stag Hunt game, each player has two options: they can choose to hunt either a hare (H) or a stag (S).
* If a player chooses to hunt a hare, they will always catch one and receive a payoff of $h>0$.
* However, if both players choose to hunt a stag, they will both receive a higher payoff of $s>h$. Hunting stag is potentially more profitable than hunting hare, but it requires a coordinated effort from both players.

```python
import numpy as np
import nashpy as nash
p1=np.array([[4,2],[6,1]])
p2=np.array([[4,6],[2,1]])
game=nash.Game(p1,p2)
equilibria = game.support_enumeration()
for eq in equilibria:
  print(eq)
```

There are three Nash equilibria in total, two of which are mixed strategies and one of which is a pure strategy.

* Pure strategy equilibrium: (Stag, Stag)
  Both players hunt the stag and receive a payoff of 4.

* Mixed strategy equilibrium: (Stag with probability 2/3, Hare with probability 1/3), (Stag with probability 2/3, Hare with probability 1/3)
Each player plays stag with probability 2/3 and hare with probability 1/3. In this equilibrium, both players receive a payoff of 4/3.

* Mixed strategy equilibrium: (Hare with probability 1, Stag with probability 0), (Hare with probability 1, Stag with probability 0)
Each player plays hare with probability 1. In this equilibrium, both players receive a payoff of 1.

Therefore, there is one pure strategy equilibrium and two mixed strategy equilibria. In the pure strategy equilibrium, both players hunt the stag. In the mixed strategy equilibria, each player randomizes between hunting the stag and hunting the hare.

# Extensive Form Game: Ultimatum Game (Güth et al 1982, 367-388)
* In the Ultimatum game, there are two players: the proposer and the responder.

* The proposer is given a sum of money and must suggest how to split it between themselves and the responder.

* If the responder agrees with the proposal, then the money is split as suggested.

* However, if the responder declines the proposal, then neither player receives any money.
